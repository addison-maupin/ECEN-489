```python
from numba import jit

@jit
def sum_of_cubes():
    finalArray = []
    for i in range(-1000, 1001):
        for j in range(-1000, i):
            for k in range(-1000, j):
                N = i**3 + j**3 + k**3
                if 0 <= N <= 100:
                    finalArray.append([N,i,j,k])
    return finalArray 
                    
def main():
    result = sum_of_cubes()
    result.sort()
    for i in range(50, 101):
        print("N =",i,":")
        for j in range(0, len(result)):
            if result[j][0] == i: 
                print("   ", result[j][1], result[j][2], result[j][3])
        print("\n")

if __name__ == "__main__":
    main()
```
